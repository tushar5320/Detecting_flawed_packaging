```
"""
data augumentation to get more data as the images of each type are few in numbers

"""

from keras.preprocessing.image import ImageDataGenerator
import cv2,os
import numpy as np
```

```
"""make a dictionary where images are stored as values and their labels are storedas keys

eg :- {0-> fraud package image
      1-> non-fraud image
      }
"""
```

```
#loading image
fraud_img='/content/drive/My Drive/tushar/POSITIVE.jpg'    #yaha daalo path
non_fraud_img='/content/drive/My Drive/tushar/negative.jpg'    #yaha daalo path

dic={0:None,1:None}

fraud=cv2.imread(fraud_img)
non_fraud=cv2.imread(non_fraud_img)

dic[0]=fraud
dic[1]=non_fraud
```

```
print(fraud)
```

```
from keras.preprocessing.image import ImageDataGenerator

for k in dic:
  images=np.array(dic[k])
  images=cv2.resize(images,(120,120),interpolation=cv2.INTER_AREA)
  images=images.reshape(1,120,120,3)

  aug = ImageDataGenerator(
    rotation_range=30,
    zoom_range=0.15,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.15,
    fill_mode="nearest")
  dir_path='/content/drive/My Drive/tushar/aug/{}'.format(k)
  print(dir_path)
  os.mkdir(dir_path)
  aug.fit(images)

  for x, val in zip(aug.flow(images,save_to_dir=dir_path,save_prefix='aug',save_format='jpg'),range(200)) :   #change this 200 to get more images
      pass

#images will be saved
```
